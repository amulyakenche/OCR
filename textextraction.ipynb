```python
import matplotlib.pyplot as plt
import PIL
import pytesseract
import re
%matplotlib inline
```

```python
img = PIL.Image.open(r"F:\elonmusk.jpg")
plt.imshow(img)
```

```python
# config
pytesseract.pytesseract.tesseract_cmd = 'C:/Program Files/Tesseract-OCR/tesseract'
TESSDATA_PREFIX = 'C:/Program Files/Tesseract-OCR'
```

```python
text_data = pytesseract.image_to_string(img.convert('RGB'), lang='eng')
```

```python
print(text_data)
```

```python
lines = text_data.split('\n')
for line in lines:
    # Check if the line starts with "Name:"
    if line.startswith("Name:"):
        # Extract the name by splitting the line at "Name:" and taking the second part
        name = line.split("Name:")[1].strip()
        break

# Print the name if found, otherwise print "No name found."
if name:
    print(name)
else:
    print("No name found.")
```
